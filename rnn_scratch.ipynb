```python
import numpy as np
import tensorflow as tf
from tqdm import tqdm
```

```python
n_nuerons=10
n_instance=100
n_feautures=6
batch_size=10
len_sequence=7
```

```python
x_t=tf.random.uniform((n_instance,len_sequence,n_feautures))
y_true = tf.random.uniform((n_instance,n_nuerons))
```

```python
W_x = tf.random.uniform((n_feautures, n_nuerons))
W_y = tf.random.uniform((n_nuerons, n_nuerons))
W=tf.Variable(tf.concat([W_x,W_y],axis=0))
b=tf.Variable(tf.random.uniform((1,n_nuerons)))
```

```python
def rnn_computation(x_t,y_t_1):
    # print(x_t.shape,y_t_1.shape)
    input_matrix=tf.concat([x_t,y_t_1],axis=1)
    z=tf.matmul(input_matrix,W) + b
    return tf.tanh(z)
```

```python
def forward_pass(x_t):
    h=x_t.shape[0]
    y_t_1=tf.ones(shape=(h,n_nuerons))
    for i in range(len_sequence):
        feed_to_network=x_t[:,i,:]
        y_t_1=rnn_computation(feed_to_network,y_t_1)
        # print(y_t_1)
    return y_t_1
```

```python

def compute_loss(ops,y_true):
    loss_fn = tf.keras.losses.CategoricalCrossentropy(from_logits=True)
    loss=loss_fn(ops,y_true)
    return loss
    
```

```python

n_epochs=10
n_iterations=int(n_instance/batch_size)
optimizer = tf.optimizers.Adam(learning_rate=1.0)
for epoch in range(n_epochs):
    iteration = 0  
    epoch_loss = 0

    with tqdm(total=n_iterations, desc=f"Epoch {epoch+1}/{n_epochs}", unit="batch") as pbar:
        for i in range(n_iterations):
            x_batch = x_t[iteration:iteration + batch_size]

            with tf.GradientTape() as tape:
                ops = forward_pass(x_batch)
                
                loss = compute_loss(ops, y_true[iteration:iteration + batch_size])
                

            gradients = tape.gradient(loss, [W, b])
            print(gradients)
            optimizer.apply_gradients(zip(gradients, [W, b]))

            iteration += batch_size  
            epoch_loss += loss.numpy()  

            pbar.set_postfix(loss=loss.numpy())  
            pbar.update(1)  

     



```
